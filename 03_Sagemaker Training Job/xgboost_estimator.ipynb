```python
import sagemaker
from sagemaker.sklearn.estimator import SKLearn
```

```python
sagemaker_session = sagemaker.Session()
role = sagemaker.get_execution_role()
```

```python
train_s3_uri = "s3://dev-ds-clean-data-1654/clean_data.parquet"
```

```python
sklearn_estimator = SKLearn(
    entry_point="train.py",
    source_dir="code_folder",
    role=role,
    instance_count=1,
    instance_type="ml.m5.large",
    framework_version="1.2-1",
    py_version="py3",
    hyperparameters={
        "n-estimators": 300,
        "learning-rate": 0.1,
        "max-depth": 6,
        "subsample": 0.8,
        "colsample-bytree": 0.8
    }
)
```

```python
sklearn_estimator.fit({"train": train_s3_uri})
```
